```python
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split#splitting data into train and test set
from sklearn.linear_model import LinearRegression #creating a linear regression model
from sklearn.metrics import r2_score,mean_squared_error #evaluating
import seaborn as sns
```

```python
df=pd.read_csv('BostonHousing.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
sns.histplot(df['medv'])
```

```python
#splitting of dataframe
X=df.drop('medv',axis=1)#all colums expect target variable
y=df['medv']#target variable
```

```python
import matplotlib.pyplot as plt
corelation=df.corr().round(2)
plt.figure(figsize=(16,10))
sns.heatmap(data=corelation,annot=True ,linewidth=0.5)
plt.show()#to visualize complex data and correlation between different feature and target variable
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)#splitting of data into 80% training and 20% testing
```

```python
model=LinearRegression()#initializing linear regression model
```

```python
model.fit(X_train,y_train)#we train the regression model using training data
```

```python
y_pred=model.predict(X_test) #we use trained model to predict on testing data
```

```python
mse=mean_squared_error(y_test,y_pred)#average square difference between actual and predicted
r=r2_score(y_test,y_pred)#proportion of variance of target variable from feature
mse,r
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(y_test,y_test,c='Blue',label='actual price')
plt.scatter(y_test,y_pred,c='red',label='predicted price')
plt.xlabel('actual price')
plt.ylabel('predicted price')
plt.show()
```
